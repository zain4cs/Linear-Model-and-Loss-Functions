```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Chota Dataset (Linear Relationship dikhane ke liye)
data = {
    'Study Hours': [1, 3, 4, 6, 8],      # X - Input Feature
    'Pass Chances': [0.2, 0.4, 0.5, 0.7, 0.9] # Y - Target (Pass hone ki probability)
}
df_small = pd.DataFrame(data)

# X aur Y variables tayyar karna (Zaroori shape mein)
X = df_small[['Study Hours']] # Input (2D array chahiye)
Y = df_small['Pass Chances'] # Target (1D array)

print("--- Small Dataset ---")
print(df_small)
```

```python
# Model banana aur train karna
model = LinearRegression()
model.fit(X, Y)

# Intercept aur Coefficient nikalna
intercept_c = model.intercept_
slope_m = model.coef_[0] # Coefficient single value hai

print("\n--- Model Results ---")
print(f"Intercept (c): {intercept_c:.2f}")
print(f"Coefficient (m): {slope_m:.2f}")

# Prediction line ka equation
print(f"Equation: Y_hat = {slope_m:.2f}X + {intercept_c:.2f}")
```

```python
# Prediction line bana rahe hain
Y_pred = model.predict(X)

# Plotting shuru karte hain
plt.figure(figsize=(8, 5))

# Asal Data Points (Scatter Plot)
plt.scatter(X, Y, color='blue', label='Actual Data Points')

# Regression Line (Prediction Line)
plt.plot(X, Y_pred, color='red', label='Regression Line')

# Y-axis par Intercept ko highlight karna
plt.axvline(x=0, color='gray', linestyle='--') # X=0 ki line
plt.axhline(y=intercept_c, color='red', linestyle=':', linewidth=2) # Intercept ki value
plt.plot(0, intercept_c, 'ro') # Intercept point ko circle karna

plt.title('Visualizing Linear Regression (Study Hours vs Pass Chances)')
plt.xlabel('Study Hours (X)')
plt.ylabel('Pass Chances (Y)')
plt.legend()
plt.grid(True)
plt.show()

# Conclusion based on visualization:
print("\n--- Visual Summary ---")
print(f"1. Intercept ({intercept_c:.2f}): Line Y-axis ko yahan kaat rahi hai (X=0 par).")
print(f"2. Coefficient ({slope_m:.2f}): Line ki Dhalan positive hai, matlab Study Hours badhne se Pass Chances tezi se badh rahe hain.")
```

```python
class linear:

  def __int__(self):
    self.m = None
    self.b = None

  def fit(self, x_train, y_train):

    num = 0
    den = 0

    for i in range(x_train.shape[0]):
      num = num + ((x_train[i] - x_train.mean())*(y_train[i]-y_train.mean()))    # (x-x.mean()) * (y - y.mean())
      den = den+ ((x_train[i] - x_train.mean())*(x_train[i]-x_train.mean()))     # (x - x.mean)2

    self.m = num/den                                                              # m = (x-x)(y-y) / (x-x)2
    self.b = y_train.mean() - (self.m * x_train.mean())                           # b = (y.mean() - m x.mean()) ||-(y - mx)
    print(self.m)
    print(self.b)

  def predict(self, x_test):
    return self.m * x_test + self.b
```

```python
x = df[['Final_Exam_Score','Past_Exam_Scores']]
y = df['Pass_Fail']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=3)
```

```python
class lr:
  def __int__(self):
    self.coef_=  None
    self.intercept_ = None

  def fit(self, x_train, x_test):
    x_train = np.insert(x_train, 0,1, axis=1)

    bates = np.linalg.inv(np.dot(x_train.T, x_train)).dot(x_train.T).dot(y_train)

    self.intercept_ = bates[0]
    self.coef_ = bates[1:]

  def predict(self, x_test):
    y_pred = np.dot(x_test, self.coef_) + self.intercept_
    return y_pred
```
